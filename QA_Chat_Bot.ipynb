# Importing necessary libraries

```python
import os
import langchain
import pinecone
from langchain.document_loaders import PyPDFLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.vectorstores import Pinecone
```

## Loading the Document

```python
#To Read the Document
file_loader=PyPDFLoader("/content/Fundamentals of Business (complete).pdf")
documents=file_loader.load()
```

```python
len(documents)
```

##  Convert the data to chunks and then embedding using a embedding model

```python
text_splitter=RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=50)
doc=text_splitter.split_documents(documents)
```

```python
len(doc)
```

```python
embeddings=HuggingFaceEmbeddings()
```

```python
vectors = embeddings.embed_query("how are you?")
len(vectors)
```

## Set Up the Pincone serverless DB

```python
import getpass
```

```python
from pinecone.grpc import PineconeGRPC as Pinecone
from pinecone import ServerlessSpec
```

```python
os.environ['PINECONE_API_KEY'] = getpass.getpass("Enter your Pinecone_api_key: ")
```

## Initialize the database

```python
#Creating Pinecone index
pc= Pinecone(api_key=os.environ.get("PINECONE_API_KEY"))

index_name="business-index"

if index_name not in pc.list_indexes().names():
  pc.create_index(
      name="business-index",
      dimension=768,
      metric="cosine",
      spec=ServerlessSpec(
          cloud="aws",
          region="us-east-1"
      )
  )
```

## Add embeddings to the database

```python
from langchain_pinecone import PineconeVectorStore

docsearch = PineconeVectorStore.from_documents(
    documents=doc,
    index_name=index_name,
    embedding=embeddings,
    namespace="wondervector5000"
)
```

## Train the LLM: I have used Mistral Model from Huggingface as i dont have subscription for openAI hope you will consider

```python
os.environ['HUGGINGFACE_ACCESS_TOKEN']=getpass.getpass("Enter your Huggingface_Token to proceed further : ")
```

```python
#setup LLM model
from langchain_huggingface import llms, HuggingFaceEndpoint


# Define the repo ID and connect to Mixtral model on Huggingface
repo_id = "mistralai/Mistral-7B-Instruct-v0.2"
llm = HuggingFaceEndpoint(
  repo_id=repo_id,
  huggingfacehub_api_token=os.getenv('HUGGINGFACE_ACCESS_TOKEN')
)
```

## Optimizing Technique 1 - Prompting




```python
from langchain_core.prompts import PromptTemplate

template = """
You are a Business Advisor. These Human will ask you a questions about Business
Use following piece of context to answer the question.
If you don't know the answer, just say you don't know.
Keep the answer more than 4 sentences and concise.

Context: {context}
Question: {question}
Answer:

"""

prompt = PromptTemplate(
  template=template,
  input_variables=["context", "question"]
)
```

```python
from langchain.chains import RetrievalQA
qa = RetrievalQA.from_chain_type(
    llm=llm,
    chain_type="stuff",
    retriever=docsearch.as_retriever()
)
```

```python
# Ask a question and see the response
question = "What are the fundamentals of the business?"
response = qa(question)
print("Response from chatbot:", response)
```

## Optimizing Technique 2 - Similarity Retreive

---



```python
def retrieve_query(query,k=5):
  matching_results= docsearch.similarity_search(query,k=k)
  return matching_results
```

```python
from langchain.chains.question_answering import load_qa_chain
chain=load_qa_chain(llm,chain_type="stuff")
```

```python
def retireve_answers(query):
  doc_search=retrieve_query(query)
  print(doc_search)
  response = chain.run(input_documents=doc_search, question = query)
  return response
```

```python
our_query = "What are the key things for the business?"
answer= retireve_answers(our_query)
print(answer)
```
